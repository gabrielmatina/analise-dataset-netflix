```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("netflix_titles.csv")
```

```python
df.head()
```

## Cada um dos títulos são separados por tipos: Movie ou TV Show

```python
df_tipos = df['type'].drop_duplicates().copy()
df_tipos.tolist()
```

## Os filmes são categorizados na coluna ''listed_in"

- Cada filme pode ter mais de uma categoria

```python
df_categoria = df['listed_in'].str.split(', ').explode().tolist()

df_categorias = list(set(df_categoria))
```

## Dataframe apenas com filmes:

```python
df_filmes = df[df['type'] == 'Movie']
df_filmes.head()
```

```python
# Separar cada item da coluna 'listed_in' em uma nova linha
df_filmes_expandido = df_filmes.assign(listed_in=df_filmes['listed_in'].str.split(', ')).explode('listed_in').reset_index(drop=True)

# Imprimir o DataFrame expandido
df_filmes_expandido.head()
```

```python
df_contagem_categorias_filmes = df_filmes_expandido.groupby('listed_in')['type'].count()
df_contagem_categorias_filmes = pd.DataFrame(df_contagem_categorias_filmes)
df_contagem_categorias_filmes
```

```python
#Ordena o DataFrame por média em ordem crescente para melhorar a visualização.
# df_qtd_categorias = df_qtd_categorias.sort_values()

#Configura os eixos do gráfico.
ax = df_contagem_categorias_filmes.plot(kind='bar', color='skyblue')
ax.set_xlabel('Categorias')  # Define o rótulo do eixo x
ax.set_ylabel('Quantidade')  # Define o rótulo do eixo y
ax.set_title('Quantidade de Títulos por Categoria')  # Define o título do gráfico
```

## As categorias de cada TV Show foram separadas

```python
# Separar cada item da coluna 'listed_in' em uma nova linha

# ACRESCENTAR O COUNTRY
df_tv_shows_expandido = df_tv_show.assign(listed_in=df_tv_show['listed_in'].str.split(', ')).explode('listed_in').reset_index(drop=True)

# Imprimir o DataFrame expandido
df_tv_shows_expandido.head()
```

```python
# analisar a qtd de lancamentos por pais, soma do title
df_tvshow_country = df_tv_show[['title', 'country']].copy()

df_tvshow_country = df_tvshow_country.assign(country=df_tv_show['country'].str.split(', ')).explode('country').reset_index(drop=True)
df_tvshow_country = df_tvshow_country.groupby('country')['title'].count()
# df_tvshow_country = pd.DataFrame(df_tvshow_country['title'])
df_tvshow_country
```

```python
import pandas as pd

# Seu código para contar a quantidade de lançamentos por país
df_tvshow_country = df_tv_show[['title', 'country']].copy()

df_tvshow_country = df_tvshow_country.assign(country=df_tvshow_country['country'].str.split(', ')).explode('country').reset_index(drop=True)
df_tvshow_country = df_tvshow_country.groupby('country')['title'].count()

# Transformando a série em um DataFrame
df_tvshow_country = pd.DataFrame(df_tvshow_country).reset_index()
df_tvshow_country.columns = ['Country', 'Number of TV Shows']  # Renomeando as colunas

# Mostrando o DataFrame
df_tvshow_country
```

```python

# Seu código para contar a quantidade de lançamentos por país
df_tvshow_country = df_tv_show[['title', 'country']].copy()

df_tvshow_country = df_tvshow_country.assign(country=df_tvshow_country['country'].str.split(', ')).explode('country').reset_index(drop=True)
df_tvshow_country = df_tvshow_country.groupby('country')['title'].count()

# Transformando a série em um DataFrame
df_tvshow_country = pd.DataFrame(df_tvshow_country).reset_index()
df_tvshow_country.columns = ['Country', 'Number of TV Shows']  # Renomeando as colunas

# Ordenando os dados por 'Number of TV Shows' (opcional)
df_tvshow_country = df_tvshow_country.sort_values(by='Number of TV Shows', ascending=False)

# Criando o gráfico de barras
plt.figure(figsize=(10, 6))  # Define o tamanho da figura

# Plotagem do gráfico de barras
plt.bar(df_tvshow_country['Country'], df_tvshow_country['Number of TV Shows'], color='skyblue')

# Adicionando rótulos e título
plt.xlabel('Country', fontsize=14)
plt.ylabel('Number of TV Shows', fontsize=14)
plt.title('Number of TV Shows by Country', fontsize=16)
plt.xticks(rotation=90, ha='right')  # Rotaciona os rótulos do eixo x para facilitar a leitura

plt.xlim(0.5, df_tvshow_country['Country'].max())

# Mostrando o gráfico
plt.tight_layout()  # Ajusta o layout para garantir que nada seja cortado
plt.show()
```
